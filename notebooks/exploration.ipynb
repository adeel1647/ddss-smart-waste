```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
```

```python
DATASET_DIR = "../data/images"
classes = ["cardboard", "glass", "metal", "paper", "plastic", "trash"]
```

```python
plt.figure(figsize=(15, 10))

img_count = 1

for cls in classes:
    class_dir = os.path.join(DATASET_DIR, cls)
    images = os.listdir(class_dir)
    
    sample_images = random.sample(images, 5)
    
    for img_name in sample_images:
        img_path = os.path.join(class_dir, img_name)
        img = Image.open(img_path)
        
        plt.subplot(len(classes), 5, img_count)
        plt.imshow(img)
        plt.axis("off")
        
        if img_count % 5 == 1:
            plt.ylabel(cls, fontsize=12)
        
        img_count += 1

plt.suptitle("Sample Images from TrashNet Dataset", fontsize=16)
plt.tight_layout()
plt.show()
```

## Dataset Visual Inspection

To verify dataset integrity and correct class labelling, a random subset of images
from each waste category was visually inspected. This step ensures that the dataset
structure aligns with the expected input format for Keras-based image classification
pipelines and that class labels are consistent with visual content.
